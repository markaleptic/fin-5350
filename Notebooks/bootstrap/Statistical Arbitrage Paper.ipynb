## Fundamental Questions of the Paper

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Performance of statistical arbitrage in petroleum futures markets, by Amir H. Alizadeh and Nikos K. Nomikos, is a paper which more fully cements the power of computational finance. Building on the research and models of previous papers, they explore the combination of fundamental and technical trading strategies in the petroleum markets. In addition, they take an additional step by using a stationary bootstrap to determine if common trading strategies are useful in analyzing and issuing trading signals for non-historical data. Bootstrapping proves essential to confirming the viability of the trading strategy. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In the words of the authors, the paper’s objective “...is to investigate the relationship between the different pairs of petroleum and petroleum products futures spreads and utilize these linkages to establish trading strategies that make use of statistical arbitrage trading opportunities.” The futures that they examine are traded on the NYMEX and the ICE; they create different pairs using WTI crude oil and NYMEX heating oil from the NYMEX and ICE Brent crude oil and ICE gas oil. By examining the spread of the pairs they create and analyzing trading strategies triggered by the spread, they hope to fill a gap in oil market research by providing evidence of trading opportunities through statistical arbitrage.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;To provide this evidence, the authors seek to answer two central questions. The first question, if these technical strategies apply to data from “out-of-sample tests, before making inferences”, focuses on the viability of the strategies in other scenarios. The strategies analyzed and proposed in the past were often modeled only on historical data, leaving little question of their functionality (since they would only be recommended having made a statistically significant improvement to returns). Data snooping occurs “when a data set is used more than once for data selection and inference purposes”. By using the historical data repeatedly to evaluate models and trading strategies, it is possible to find one that fits the historical data simply by pure probability. By using the bootstrap, which we will discuss and replicate later in this paper, the authors hoped to create random data sets that still fit the statistical properties of the historical data in order to improve the usefulness and fit of the strategies they investigate to future real-world uses.  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The second main question they hope to answer is also quite relevant. They seek to determine if the profits produced by different strategies significantly greater than other benchmark strategies. Previous papers determined that significant returns can be generated from technical and fundamental trading strategies, but the data snooping concern related to previous studies encouraged them to reexamine the profitability of the strategies on different data. The setup the authors use, with the comparison of different combinations and their spreads, “enables [them] to assess whether the dynamic strategy of frequent rebalancing according to signals specified by the behavior of price differentials among petroleum futures is superior to static strategies”. Their analysis does indicate that the strategies are superior to static strategies, even after reducing the possibility of data snooping.

## Methodology

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The trading methodology used by the authors was influenced by a paper by Dunis which modeled and forecasted spreads to analyze various strategies. Referencing the daily closing prices of two petroleum products, he tested five different trading models. Dunis found that the best models proved to be the autoregressive moving average (ARMA) and moving average (MA). The question lingers as to whether the strategies would still produce similar results if data snooping were not included, and the authors seek to help answer a piece of this question by using the MA model on different futures pairs with historic and stationary bootstrapped data sets. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The statistical methodology used by the authors utilizes a vector error correction model, or VECM. The statistical setup can be referenced in their paper, but the main goal is to examine the potential for causal relationships between pairs of contracts. They use the VECM developed by Johansen to determine this causal relationship. The existence of a stationary relationship between different futures pairs is first examined, by using the lamda max and lambda trace statistics to test the rank of the second coefficient matrix. If it is full rank, then the futures prices are I(0), then the correct strategy to estimate the model is a vector autoregression model in levels. If it has a reduced rank, there is at least one cointegrating vector and the coefficient matrix can be decomposed to see the speed of convergence to the long-run state. The authors explain the importance of cointegration by saying “…if the pairs of futures prices are cointegrated, then causality must exist in at least one direction”. If this cointegration exists, it means that the future prices of one contract can be better predicted using the past prices of the other contract, along with other historical information. The VECM proved to be the most effective model with I(1) variables, and the Johansen procedure provides a more efficient estimate of cointegration as well as better performance with non-normality and heteroskedastic disturbances. The model lets the authors look at the comovement and linkage between two oil futures. The model also expresses the short-run price dynamics with the lagged cross-market returns and long-run price trends in the cointegration vector. From these pieces, the error correction term provides the spread between logfutures prices and also leads to the average long-run spread. These insights are important in establishing the model, as well as determining the viability of the bootstrap.

## Model Assumptions

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The model assumptions used in the paper will be mentioned, but not discussed in great detail. They are as follows:
1.	Since the contracts’ expiration dates are not matching, it is assumed that the investor will roll over to the front month pair of contracts on the first day of the last trading month.
2.	If two prices are cointegrated, causality must exist in at least one direction. Prices are determined by supply and demand in different ways for each contract
3.	Since transaction fees would be a significant cost for more active trading strategies, a transaction fee of .2% per round trip trade is used
4.	Commission charges are assumed to be negligible
5.	Differentials of the futures prices under study are stationary, for bootstrap to be viable

## Trading Strategies

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;As mentioned before, the authors focus on the MA trading strategy. The strategy is based on the comparison of one fast (short) and one slow (long) MA of the spread of futures prices. One strategy that they use is to compare a three-month MA of the spread with a one-week MA of the same series. From these positions, the short-term and long-term averages crossing paths generate buy/sell or long/short signals to execute a trade. The authors explain that “the positive difference between three-month and one-month indicate a buy decision. If the three-month MA is greater than the one-week MA, it can be interpreted as the series being below its long-run average. This indicated that the futures prices of the one commodity are undervalued relative to the other.” From examining the spread generated by different pairings and time lengths, they look to generate the buy/sell signals for trading.

## Test the Trading Strategies with Empirical Data

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The authors first tested the trading strategies using empirical market data. They used the weekly futures prices of four energy commodities: NYMEX WTI (West Texas Intermediate) crude oil, NYMEX heating oil, ICE Brent crude oil and ICE gas oil. The pairs created for spread analysis and trading were CBt – GOt, CLt – GOt, GOt – HOt, CLt – HOt, CBt – CLt, and CBt – HOt. Data was from August 9th, 1989 to September 20, 2006, providing a total of 894 weekly observations from Wednesday prices, or from Tuesday prices if Wednesday fell on a holiday when the markets were closed. They utilized contracts with a variety of lengths, and combined them to create the spread pairings examined in the study by using a one, two, three, and four-month lag. After looking at the data, the authors discovered that there was a difference in the descriptive statistics of the futures prices and the spread series. The futures prices for the crude and the more refined petroleum are difference stationary, while the spread series were stationary, which led them to conclude that the VECM specification was the right method to determine the “relationships and mean-reverting properties of the spread series”. Also of note, the spread series were serially correlated, heteroskedastic, and non-normal.

## Empirical Results

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;After examining the information in several separate ways, the authors reached a variety of conclusions that support the belief that there are profitable trading opportunities in the oil markets. Johansen cointegration tests indicated that the oil futures prices stand in a long-run relationship with each other. This implies that changes in the “relationship” of the prices, or their closeness to each other, shows a trading opportunity since the change will correct back towards the long-run mean. When combined with correction of the standard errors for heteroskedasticity and serial correlation of all regressions, they found that they could determine implied responses following different deviations from long-run relationships along with other information.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;They discovered several different trends. They found that crude oil tends to lead the information discovery process in pairings with either NYMEX or ICE crude oil futures paired with ICE gas oil and NYMEX heating oil for almost all maturities. The exceptions were the four-month ICE Brent (NYMEX heating oil and NYMEX WTI) heating oil pairs which did not indicate a significant lead-lag relationship. Additionally, their estimates of error correction coefficients indicated that heating and gas oil prices tend to adjust to equilibrium, where the crude prices are not as responsive to departures from the long-run mean. The performance of the MA trading rules help to illustrate some of these trends.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;They used four MA series differentials between the log futures prices to create the data being examined. One fast, or short, which was called MA(1), and three slow, or long, called MA(4), MA(8), MA(12). The difference between the two constructed MA series was used to generate buy/sell signals for the petroleum futures spread. The sign of the difference between the slow and the fast MA was used as the signal, with positive difference indicating a buy and a negative difference indicating a sell. Overall, for each spread analyzed an MA strategy produced higher returns than the benchmark strategy of buy and hold. Additionally, the potential for profit also increased while lowering the overall risk as measured by the standard deviation. The returns ran between -6.06% and 29.17%, compared to 6.28% - 7.47% for the benchmark strategy. The Sharpe ratios also indicated that most MA strategies outperformed the buy and hold default strategy. From this historical data, it was fairly evident that the MA strategies provided several advantages over the benchmark approach.

## Testing the Trading Strategies with Random Data from Bootstrap

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;As mentioned earlier, data snooping is a potential issue when only referencing historic data to validate models. The authors highlight this fact by saying that “using the same data set frequently for testing trading strategies may increase the probability of having satisfactory results purely due to chance or due to the use of posterior information rather than the superior ability of the trading strategies”. In other words, it is crucial to also confirm the usefulness of a trading strategy by applying it to different data sets. To allow for reasonable assumptions of the distribution of statistics of interest in generating new random data, the empirical distribution of the data is used to model the new distribution of the data. This helps to approximate the characteristics of the sampling distribution of the underlying statistic regardless of the complexity of the data. In order to maintain the statistical properties of the original data set and yield consistent results, the authors used a stationary bootstrap in place of an ordinary bootstrap.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The stationary bootstrap is primarily used to regenerate random paths which futures prices could have followed over the empirical sample period. The authors begin by bootstrapping the logdifference series, which are then transformed back into levels to make the spreads for triggering the buy/sell decisions for the trading strategies. The benchmark models they use are the buy and hold trading strategy where the trader is always long in one of the two legs of the spread. They used the MA strategies as well as the two buy and hold strategies for each of the 1,000 bootstrapped series. By doing so they are able to create a series of mean returns and Sharpe ratios, allowing them to compare the performance of the models to sample data with similar statistical properties as the empirical data. The mean annualized returns and average Sharpe ratios for the bootstrapped series were quite similar to that of the empirical data. The comparative performance between the different MA strategies is also similar. They specifically mention that the higher returns come from the three spreads that include ICE gas oil in one leg, and that the worst performing strategy is the four-month NYMEX crude-heating oil spread. Overall, the MA-based trading rules outperform the more stable buy/hold in terms of increasing average returns and the Sharpe ratio. Of all of the MA strategies, the highest Sharpe ratio is achieved by the ICE gas oil – NYMEX heating oil spread, with more than an 8x increase compared with the more static, buy-hold strategies.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Statistical p-value tests demonstrated that the excess Sharpe ratio and returns from the MA strategies are statistically significant from zero. From these tests, it can be said that the MA strategies can provide meaningful increases in returns over ordinary static strategies. For the one, two, and three–month cases the MA strategies provided significantly higher returns in four out of six spreads, and three out of six for the four-month spreads. The majority of the spreads were significant at the 1% level, with all of them significant at the 10% level. From these data, it is clear that there are significant benefits achievable by using spread variations to find mispricing in the oil futures market. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;However, transaction costs affect the returns and may vary due to factors such as the type of trader and the market involved. A sensitivity analysis was done, comparing the .2% assumption to a range of .3%, .4%, and .5% transaction costs for all maturities of each trading strategy. Even with .5% transaction costs, there were three of the six spreads which produced significantly higher Sharpe ratios, amounting to returns 2 to 4 times higher than for the traditional buy and hold strategies. This implies that for these three spreads, the tendency to move back to the mean long-term spread isn’t fully impounded in market expectations, creating the arbitrage opportunities. One other spread, the one-month intercrude spread, also produced significantly higher Sharpe rations at the .3% and .4% transaction costs. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The last, and vital, question is why some spreads perform better than others and how opportunities can be identified from the spreads. One of the main reasons comes from the Johansen cointegration tests used in the VECM. The lambda trace and lambda max statistics indicate that futures prices move together in the long run, but that they show a different behavior in the short run. In the long run, the movements are generally due to the mutual exposure to the world oil market. Because the assets are so correlated, it is possible to find arbitrage opportunities by taking opposite positions simultaneously and riding the market correction. Differences of spread prices tend to occur generally in the short run, and are normally attributed to seasonal changes, more regional and country-specific supply and demand, and any new information that enters the market. If there is a high correlation between the legs of a pair in the long run and a lower correlation in the short-run, then it may be more likely that there is the potential for arbitrage since there may be short-run divergences in price that will correct in the future. It is also important to account for the volatility of prices and returns, since it influences how much the futures pair will drift from the long-run relationship. With these two factors, it is expected that higher long-run correlations combined with lower short-run correlations yield the best outcome. The volatility is market-specific, however, and should be closely examined to better understand how the futures spread will diverge. In short, “…long- and short-run correlation justifies pairs trading and long- and short-run volatility includes information on the potential profitability of the spread under study, [because] as the spread deviates from the long-run equilibrium, the probability of convergence [and achieving higher profits] is increased”. The further apart the pair drifts, the more potential profits from introducing a simultaneous long/short trading position. The examination of these indicators, however, merely indicates the potential for profit, as transaction costs still must be covered and it must beat the simple benchmark strategy, and should be used in combination with other fundamental and technical analysis.

## Conclusions

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The paper provided significant insights into the oil and petroleum futures market. After using the VECM to discover the main market drivers and causality, they used the discovered cointegration relationship to develop the trading strategy on petroleum futures spreads and their variation from their long-run mean. The MA of these pairs was used to create the buy/sell signals for the trading strategy, which proved to provide statistically significant increases in Sharpe ratios and returns for all simple MA strategies examined over the benchmark buy and hold strategy, subject to transaction costs. This proved true for both historical data and for bootstrapped data which shared the same statistical properties of the historical data set. From the relationships between futures contracts, and the long- and short-run correlation and volatility of the spreads, arbitrage opportunities can be identified in the market. An investor can rest assured that the trading strategy can be used profitably in the real world, because the true power of these MA trading strategies is confirmed to be robust with out-of-sample testing. Overall, their results indicated that the relationships of some petroleum futures can be used to create profitable trading strategies based on market timing, and that simple MA indicators are able to take advantage of the trading opportunities to create returns greater than comparable benchmarks.

<h2><center>Boostrap Replication</center></h2>

```python
import numpy as np


def stationary_bootstrap(data_vec_to_resapmle, q = 0.1, replications = 1000):
    """
    Algorithm to implement stationary boostrap resampling technique as described by Politis and Romano (1994), 
    Appendeix C of Sullivan et all (1999), and most specifically from Appendex 1 of Alizadeh and Nomikos (2008).
    Inputs:
        data_vec_to_resapmle    Original data set used for resampling
        q                       Probability that the next observation in the resampled array is randomly
                                selected from the original observations
        
        replications            Number of resampled arrays to return
    Return:
        1,000 (or however many designated) numpy array's resampled from data_vec_to_resapmle
    """

    vec_len = len(data_vec_to_resapmle)
    end_of_vec_index_adj = vec_len - 1
    resample_vec = np.zeros(shape=(replications, vec_len), dtype=float)
 
    for i in range(replications):
        index_vec = np.zeros(vec_len, dtype=int)
        starting_pos = np.random.randint(0, vec_len, dtype=int)
        index_vec[0] = starting_pos

        std_unifs = np.random.normal(0.0, 1.0, vec_len)

        for j in range(1, vec_len):
            if std_unifs[j] < q:
                index_vec[j] = np.random.randint(0, vec_len, dtype=int)
            else:
                if index_vec[j - 1] == end_of_vec_index_adj:
                    index_vec[j] = 0
                else:
                    index_vec[j] = index_vec[j - 1] + 1

        resample_vec[i] = data_vec_to_resapmle[index_vec]
        index_vec = np.zeros(vec_len, dtype=int)

    return resample_vec
```
